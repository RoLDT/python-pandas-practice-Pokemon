# Filtering Data and Regex

```python
df = poke_df
```

```python
#df.loc[df["Type 1"] == "Grass"] 

#When doing more than one filter, separate each one with ()
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] >70)]

#"Drop = True" to remove the old index from the dataframe; "Inplace = True" to do it in the same dataframe
new_df.reset_index(drop=True, inplace = True)
new_df
```

```python
#Filter out the Mega Evolutions

#re are regular expressions; can be used to filter data on textual patterns
import re

#The ~ symbol filters those pokemon out
#df.loc[~df["Name"].str.contains("Mega")]

#Used the regex to find both Fire and Grass types; used the flag re.I to ignore capitalization
#df.loc[df["Type 1"].str.contains("Fire|Grass",flags=re.I , regex=True)]

#Other Example. Looks for names that starts with pi with the hat (^), [a-z] means that the next set of letters
#can go from letter a thru z,and the * specifies "Zero or more"
df.loc[df["Name"].str.contains("^pi[a-z]*",flags=re.I , regex=True)]
```
